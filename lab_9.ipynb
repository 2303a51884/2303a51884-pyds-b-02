```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
data = fetch_california_housing(as_frame=True)
df = data.frame
housing_data = df[['MedInc', 'AveRooms', 'AveOccup', 'MedHouseVal']]
housing_data = housing_data.rename(columns={
    'MedInc': 'Median_Income',
    'AveRooms': 'Average_Rooms',
    'AveOccup': 'Average_Occupancy',
    'MedHouseVal': 'Median_House_Value'
})
sns.set(style="whitegrid", palette="coolwarm", font_scale=1.1)
pair_plot = sns.pairplot(
    housing_data,
    diag_kind='kde',
    corner=True,
    plot_kws={'alpha': 0.6, 's': 40, 'edgecolor': 'black'}
)
pair_plot.fig.suptitle(
    "Housing Data Analysis: Relationships Among Price, Area, and Rooms",
    fontsize=14, fontweight='bold', y=1.02
)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
titanic = sns.load_dataset("titanic")
titanic_data = titanic[['age', 'fare', 'pclass', 'survived']].dropna()
sns.set(style="whitegrid", palette="coolwarm", font_scale=1.1)
pair_plot = sns.pairplot(
    titanic_data,
    hue='survived',
    vars=['age', 'fare', 'pclass'],
    diag_kind='kde',
    corner=True,
    plot_kws={'alpha': 0.6, 's': 40, 'edgecolor': 'black'}
)
pair_plot.fig.suptitle(
    "Titanic Passenger Patterns: Relationships among Age, Fare, and Class",
    fontsize=14, fontweight='bold', y=1.02
)
plt.show()
```
