```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.DataFrame({
    "Country": ["China", "India", "United States", "Indonesia", "Pakistan"],
    "Population": [1411778724, 1393409038, 331893745, 273523621, 220892331]
})
top5 = data.sort_values(by="Population", ascending=False)
plt.figure(figsize=(8,6))
bars = plt.bar(top5["Country"], top5["Population"], color="skyblue", edgecolor="black")

plt.xlabel("Country", fontsize=12)
plt.ylabel("Population (in billions)", fontsize=12)
plt.title("Population of Top 5 Countries", fontsize=14)
def human_format(num):
    if num >= 1_000_000_000:
        return f'{num/1_000_000_000:.1f}B'
    elif num >= 1_000_000:
        return f'{num/1_000_000:.1f}M'
    else:
        return str(num)
plt.gca().set_yticklabels([human_format(y) for y in plt.gca().get_yticks()])
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
             human_format(bar.get_height()),
             ha='center', va='bottom', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.DataFrame({
    "Genre": ["Drama", "Comedy", "Action", "Thriller", "Romance",
              "Horror", "Animation", "Documentary", "Fantasy", "Sci-Fi"],
    "Movie Count": [2500, 1800, 1600, 1200, 950, 800, 700, 400, 350, 300]
})
data = data.sort_values(by="Movie Count", ascending=False)
plt.figure(figsize=(10,6))
bars = plt.bar(data["Genre"], data["Movie Count"], color="skyblue", edgecolor="black")
plt.xlabel("Genre", fontsize=12)
plt.ylabel("Number of Movies", fontsize=12)
plt.title("Number of Movies Produced by Genre", fontsize=14)
plt.xticks(rotation=45, ha="right")
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
             str(bar.get_height()), ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
